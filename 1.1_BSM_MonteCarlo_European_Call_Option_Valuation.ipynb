```python
#Book: Python for Finance by Dr. Yves J Hilpisch
#Chapter 1: Why Python for Finance? 
#Exercise 1.1: Monte Carlo Valuation of European Call Options
#Black-Scholes-Merton Model for Options Pricing Model 
#Description: 
#Black-Scholes-Merton Model explanation at Corporate Finance Institue

import math 
import numpy as np
import matplotlib.pyplot as plt

# Initial Index Level 
S0 = 100 
#Strike Price
K = 105 
#Maturity (1 year)
T = 1.0
#Riskless Short Rate
r = 0.05 
#Volatility
sigma = 0.2

#Number of simulations
I = 100000

np.random.seed(1000)

#Valuation Algorithm 
z = np.random.standard_normal(I)

#Index Values at Maturity
ST = S0 * np.exp((r - sigma ** 2/2) * T + sigma * np.sqrt(T) * z )

#Payoff at Maturity
hT = np.maximum(ST - K, 0)

#Monte Carlo Simulator
C0 = math.exp( - r * T) * np.mean(hT)

#Result Output
print('Value of the European Call Option %5.3f.' %C0)

# Plot histogram of option payoffs
plt.figure(figsize=(10, 6))
plt.hist(hT, bins=50, density=True, alpha=0.6, color='g')
plt.title('Distribution of Option Payoffs')
plt.xlabel('Option Payoff')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
